# FUNCIONES DE PYTHON

## INTRODUCCIÓN

## 1. FUNCIÓN "GREET()"

```python
def greet():
  print("Hello, world")
greet()
```

## 2. FUNCIÓN SHOW_NAME

```python
def show_name(name):
    print("Hello", name)
show_name("Thaliana")
```

## 3. FUNCIÓN ABB( A, B)

```python
def add(a, b):
    return a + b

print(add(5, 3))
```

## 4. FUNCIÓN RESTAR (A, B)

```python
def subtract(a, b):
    return a - b
print(subtract(10, 4))
```

## 5. FUNCIÓN MULTIPLY (A ,B)

```python
def multiply(a, b):
    return a * b
print(multiply(6, 7))
```

## 6. FUNCIÓN DIVIDIR

```python
def divide(a, b):
    return a / b

print(divide(20, 5))
```

## 7. FUNCIÓN IS_EVEN(NUMERO)

```python
def is_even(number):
    return number % 2 == 0

print(is_even(8))
print(is_even(3))
```

## 8. FUNCIÓN FIND_MAX (A, B)

```python
def find_max(a, b):
    return a if a > b else b

print(find_max(12, 9))
print(find_max(0.9, 0.999))
```

## 9. FUNCIÓN SQUARE(NUMERO)

```python
def square(number):
    return number ** 2

print(square(4))
print(square(-9))
```

## 10. FUNCIÓN CUBE(NUMERO)

```python
def cube(number):
    return number ** 3

print(cube(3))
print(cube(-5))
```

## 11. FUNCIÓN COUNT_LETTERS(PALABRA)

```python
def count_letters(word):
    return len(word)

print(count_letters("Python"))
print(count_letters("Esternocleidomastoideo"))
```

## 12. FUNCIÓN CONVERT_TO_UPPER(TEXT)

```python
def convert_to_upper(text):
    return text.upper()

print(convert_to_upper("iguana"))
```

## 13. FUNCIÓN CALCULATE_AREA(LENGTH)

```python
def calculate_area(length, width):
    return length * width

print(calculate_area(5, 10))
print(calculate_area(4,3.5))
```

## 14. FUNCIÓN PROMEDIO(NUMERO)

```python
def average(numbers):
    return sum(numbers) / len(numbers)

print(average([14, 210, 1, 29]))
```

## 15. FUNCIÓN FACTORIAL(N)

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))
```
