# PYTHON LISTAS

## INTRODUCCIÓN

Las listas en PYTHON nos permiten almacenar diferentes valores y manejarlos de forma ordenada y simple,Con las listas podemos realizar operaciones como agregar, eliminar, ordenar y recorrer elementos, lo que las convierte en una herramienta fundamental para la programació.

## 1. CREAR LISTA DE FRUTAS

```python
Fruits = ["apple", "banana", "orange", "grape"]
print(Fruits)
```

## 2. PRIMER ELEMENTO

```python
numbers = [4, 8, 15, 16, 23, 42]
print(numbers[0])
```

## 3. ULTIMO ELEMENTO

```python
print(numbers[5])
```

## 4. SEGUNDO ELEMENTO Y TERCERO DE LA LISTA

```python
print(numbers[1:3])
```

## 5. CIUDADES

```python
Ciudades = ["Paris" , "Madrid", "Estambul", "Cancun", "Berna"]
print(len(Ciudades))
```

## 6. AGREGAR "HAMSTER"

```python
animals = ["dog", "cat", "rabbit"]
animals.append("hamster")
print(animals)
```

## 7. AGREGAR EN LA SEGUNDA POSICION "BIRD"

```python
animals.insert(1, "bird")
print(animals)
```

## 8. ELIMINAR VALOR "30"

```python
numbers = [10, 20, 30, 40, 50]
numbers.remove(30)
print(numbers)
```

## 9. ELIMINAR ULTIMO ELEMNTO USANDO "POP()"

```python
Letters = ["a", "b", "c", "d", "e"]
Letters.pop()
print(Letters)
```

## 10. COMPROBAR SI "MARIA" ESTA

```python
names = ["Ana", "Luis", "Maria", "Pedro"]
print("Maria" in names)
```

### 11. ORDENAR LISTA EN ORDEN ASCENDETE

```python
numbers = [5, 2, 9, 1, 7]
numbers.sort()
print(numbers)
```

## 12. ORDEN DESCENDENTE

```python
numbers = [5, 2, 9, 1, 7]
numbers.sort(reverse=True)
print(numbers)
```

## 13. NUMEROS MULTIPLICADOS POR 2

```python
numbers = [5,2,9,1,7]
doubled_numbers = [num*2 for num in numbers]
print(doubled_numbers)
```

## 14. PALABRA EN LINEAS SEPATADAS

```python
words = ["python", "list", "exercise", "practice"]
for w in words:
    print(w)
```

## 15. CALIFICACION PROMEDIO

```python
grades = [80, 75, 90, 85, 95]
average = sum(grades) / len(grades)
print(average)
```
