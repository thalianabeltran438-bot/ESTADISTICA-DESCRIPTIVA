# DICCIONARIOS PYTHON

## INTRODUCCIÓN

Los diccionarios en Python son una de las estructuras de datos más importantes y versátiles. Se utilizan para almacenar información en pares clave → valor, lo que permite organizar datos de manera clara y acceder a ellos fácilmente.



### 1. CREAR DICCIONARIO STUDENT

```python
student = {"name": "Thaliana", "age": 17, "grade": "11"}
```

Se crea un diccionario con tres pares clave-valor. "name" guarda el nombre, "age" la edad y "grade" el grado.

```python
print(student)
```

Print imprime los resultados y los muestra.

### 2. ACCEDER AL VALOR "BRAND"

```python
car = {"brand": "Toyota", "model": "Corolla", "year": 2020}
print(car["brand"])
```

Usamos car[brand] para obtener el valor asociado a la clave "brand" el cual es "Toyota".

### 3. ACEDER AL VALOR "YEAR"

```python
print(car["year"])
```

Imprime el valor asociado a "year"

### 4. AGREGAR UNA NUEVA CLAVE "CITY"

```python
person = {"name": "Maidy", "age": 25}
person["city"] = "Bogota"
print(person)
```

Los diccionarios permiten añadir nuevas claves en este caso city.

### 5. MODIFICAR VALOR DE "AGE"

```python
person = {"name": "Alice", "age": 25, "city": "Bogota"}
person["age"] = 26
person
```

Los diccionarios permiten actualizar un dato.

```python
### 6. ELIMINAR UNA CLAVE
```

```python
person = {"name": "Alice", "age": 25, "city": "Bogota"}
del person["city"]
print(person)
```

Con del se elimina la clave "city" y su valor.

### 7. OBTENER TODAS LAS CLAVES

```python
fruits = {"apple": 3, "banana": 5, "orange": 2}
print(fruits.keys())
```

Devuelve todas las claves del diccionario

### 8. OBTENER TODOS LOS VALORES

```python
print(fruits.values())
```

Devuelve los valores.

### 9. RECORRER CLAVE Y VALOR CON UN LOOP

```python
for key, value in fruits.items():
    print(key, value)
```

 Devuelve pares (clave, valor) el bucle recorre cada par e imprime ambos.




### 10. VEREFICAR SI "LUIS" EXISTE

```python
grades = {"Ana": 85, "Luis": 90, "Maria": 78}
print("Luis" in grades)
```

In revisa si "Luis" es una clave en el diccionario.

### 11. VERIFICAR SI "PEDRO" EXISTE

```python
print("pedro" in grades)
```

Como "pedro" no esta en el diccionario aparece false.

### 12. SUMAR TODOS LOS VALORES

```python
numbers = {"a": 10, "b": 20, "c": 30}
print(sum(numbers.values()))
```

Values() devuelve [10, 20, 30] y sum() suma esos valores.

### 13. CONTAR PARES CLAVE-VALOR

```python
inventory = {"pens": 10, "notebooks": 5, "erasers": 8}
print(len(inventory))
```

Cuenta cuántos pares hay en el diccionario.

### 14. RECORRER PAISES Y CAPITALES

```python
countries = {"Colombia": "Bogota", "France": "Paris", "Japan": "Tokyo"}
for country, capital in countries.items():
    print(country, "->", capital)
```

Se recorre el diccionario con .items() y se imprime cada país junto con su capital.

### 15. ENCONTRAR LA TEMPERATURA MAS ALTA

```python
temperatures = {"Monday": 22, "Tuesday": 24, "Wednesday": 21, "Thursday": 23}
print(max(temperatures.values()))
```

Devuelve la temperatura mas alta.

## CONCLUSIÓN

A través de estos ejemplos, vimos que los diccionarios permiten realizar operaciones muy variadas: desde agregar o cambiar valores, hasta recorrer todos sus elementos y aplicar cálculos sobre ellos.
